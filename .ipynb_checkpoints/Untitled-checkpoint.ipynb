```python
#10  Explain break, continue, and pass with small examples.
##continue is used to skip a specific character depend on condition ,
  ## break is use to stop the loop depend on given condition
  ## pass is nothing run in code on that condition
for char in "python":
    if char=="o":
        continue
    else:
        print(char,end="")

    
```

```python
total=0
for i in [10,20,30,40,50]:
    if total>30:
        break
    total+=i
print(total)
```

```python
for char in "python":
    if char!="o":
        print(char)
    else:
        pass
```

```python
#21. What is a nested loop? Give an example.

# nested loop is loop inside a loop
l=[[1,2,3],[4,5,6]]
for i in l:
    for j in i:
        print(j)
```

```python
#24. How can you iterate over a dictionary to print both keys and values?

d={'a':1,'b':2,'c':3}
for i,j in d.items():
    print(i,'=',j)
```

```python
#25. Write a Python program to find the sum of digits of a given number.

n=int(input("enter a number:"))
sum1=0
while n!=0:
    digit=n%10
    sum1+=digit
    n//=10
print("sum of digit of number is:",sum1)
```
